```python
import json

import folium

from sources.api.train import *
from sources.data.geo import *
```

```python
train_stations = load_station_data("../../dataset/stations/N02-20_Station.geojson")

odakyu = []
for ts in train_stations:
    if ts.line == "小田原線":
        odakyu.append(ts)
```

```python
KAIT_GEO = Coordinate(Lng=139.3399782, Lat=35.4861002)
```

```python
MAPBOX_ATTR = """© <a href="https://www.mapbox.com/about/maps/">Mapbox</a> ©
<a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a>
<strong
  ><a href="https://labs.mapbox.com/contribute/" target="_blank"
    >Improve this map</a
  ></strong
>"""
f = folium.Figure(width=600, height=400)
folium_map = folium.Map(
    location=KAIT_GEO.to_folium(),
    zoom_start=13,
    tiles="https://cyberjapandata.gsi.go.jp/xyz/std/{z}/{x}/{y}.png",
    attr=f"出典: 国土地理院ウェブサイト・地理院タイル・標準地図 {MAPBOX_ATTR}"
).add_to(f)

for ts in odakyu:
    folium.GeoJson(ts.raw_feature).add_to(folium_map)

# with open("../../dataset/stations/N02-20_Station.geojson", "r") as f:
#     j = json.load(f)
#     print(j)
# folium.GeoJson("../../dataset/stations/N02-20_Station.geojson").add_to(folium_map)

folium_map
```
